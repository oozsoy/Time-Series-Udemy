### Modeling of Stock Prices according to the random walk hypothesis
-----

This approach is based on the Markov property of the random walk where the price of a stock at a given time depends only on the previous time step. Assuming the stochastic component follows a Gaussian distribution, the stock prices than also follow a Gaussian distribution at a given time.

```python
# Number of time steps
T = 1000

# initial price

Pin = 10

# drift term

mu = 0.001
```

```python
# Set the random seed
np.random.seed(42)

# last log price 

last_lp = np.log(Pin)

# initialize log_returns and prices 
log_returns = np.zeros(T)
prices = np.zeros(T)

for t in range(T):

    # sample a random log return for each time step (stochastic component) with a std 0.01

    r = 0.01 * np.random.randn() 

    # compute the new log price as the last + drift + random stocastic component 

    lp = last_lp + mu + r 

    # store the log return and actual price 

    log_returns[t] = r + mu
    prices[t] = np.exp(lp)

    # update the last log price 

    last_lp = lp
```

```python
prices
```

```python
plt.subplots(figsize=(10,6))

plt.plot(prices)

plt.xlabel('time steps')
plt.ylabel('Stock price')

plt.title("Simulated Prices")
plt.grid()
```

```python
# Vectorize the code above to simulate the Stock Prices according to the Random Walk Hypothesis
```

```python
r_vec = 0.01 * np.random.randn(T)

mu_vec = mu * np.ones(1000)

price_vec_t = np.exp(mu_vec+r_vec) 

prices_vec = Pin * price_vec_t.cumprod(axis = 0)
```

```python
plt.subplots(figsize=(10,6))

plt.plot(prices_vec, color = "red", alpha = 0.5)

plt.xlabel('time steps')
plt.ylabel('Stock price')

plt.title("Simulated Prices")
plt.grid()
```
