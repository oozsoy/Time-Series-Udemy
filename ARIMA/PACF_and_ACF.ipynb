```python
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.graphics.tsaplots import plot_pacf, plot_acf

#For inline plotting 
%matplotlib inline                 
%config InlineBackend.figure_format = 'svg'

plt.style.use("seaborn-v0_8-dark")  
```

### PACF
---------

```python
x0 = np.random.randn(1000) #iid noise 
```

```python
plt.plot(x0)
```

```python
fig, axes = plt.subplots(figsize = (10,6))

plot_pacf(x0,ax=axes);
```

```python
# generate an AR(1) process 

x1 = [0]

for i in range(1000):

    x = 0.5 * x1[-1] + 0.1 *  np.random.randn() # try changing the sign of the linear coefficient and see the pacf plot 

    x1.append(x)

x1 = np.array(x1)


plt.plot(x1)
```

```python
# Plot below proves the AR(1) process we generated above 

fig, axes = plt.subplots(figsize = (10,6))

plot_pacf(x1,ax=axes);
```

```python
# generate an AR(2) process 

x2 = [0, 0]

for i in range(1000):

    x = 0.5 * x2[-1] - 0.3 * x2[-2] + 0.1 *  np.random.randn() # try changing the sign of the linear coefficient and see the pacf plot 

    x2.append(x)

x2 = np.array(x2)


plt.plot(x2)
```

Again the data generated with an AR(2) process does not particularly look different than the iid noise we generated before!

```python
# Plot below proves the AR(2) process we generated above 

fig, axes = plt.subplots(figsize = (10,6))

plot_pacf(x2,ax=axes);
```

There are two non-zero lags corresponding to the coefficients of the AR(2) process

```python
# generate an AR(2) process 

x5 = [0, 0, 0, 0, 0]

for i in range(1000):

    x = 0.5 * x5[-1] - 0.3 * x5[-2] - 0.6 * x5[-5] + 0.1 *  np.random.randn() # try changing the sign of the linear coefficient and see the pacf plot 

    x5.append(x)

x5 = np.array(x5)


plt.plot(x5)
```

```python
# Plot below proves the AR(5) process we generated above 

fig, axes = plt.subplots(figsize = (10,6))

plot_pacf(x5,ax=axes);
```

### ACF
--------

```python
fig, axes = plt.subplots(figsize = (10,6))

plot_acf(np.random.randn(1000),ax=axes);
```

```python
# generate MA(1) process

errors = 0.1 * np.random.randn(1000)
ma1 = []

for i in range(1000):
    
    if i >= 1:
        x = 0.5 * errors[i-1] + errors[i]
    
    else:
        x = errors[i]

    ma1.append(x)

ma1 = np.array(ma1)
```

```python
plt.plot(ma1)
```

```python
fig, axes = plt.subplots(figsize = (10,6))

plot_acf(ma1,ax=axes);
```

```python
# generate MA(2) process

errors = 0.1 * np.random.randn(1000)
ma2 = []

for i in range(1000):
    
    if i >= 2:
        x = 0.5 * errors[i-1] - 0.3 * errors[i-2] + errors[i]
    
    else:
        x = errors[i]

    ma2.append(x)

ma2 = np.array(ma2)

plt.plot(ma2)
```

```python
fig, axes = plt.subplots(figsize = (10,6))

plot_acf(ma2,ax=axes);
```
